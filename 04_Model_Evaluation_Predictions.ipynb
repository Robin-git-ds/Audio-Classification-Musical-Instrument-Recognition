```python
import os
from tqdm import tqdm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from python_speech_features import mfcc
from tensorflow import keras
import pickle
```

```python
def build_predictions(audio_dir):
  y_true=[]
  y_pred=[]
  fn_prob={}

  # #time
  # min=-70.72182492951218
  # max=69.9251863770808

  # #conv
  # min=-70.37108906636645
  # max=68.94762641128114
  step=1600

  print("Extracting features from audio")
  for fn in tqdm(os.listdir(audio_dir)):
    rate,wav=wavfile.read(audio_dir+fn)
    label=fn2classes[fn]
    c=classes.index(label)
    y_prob=[]
#config.step=1600
    for i in range(0,wav.shape[0]-step,step):
      sample=wav[i:i+1600]
      x=mfcc(sample,rate,numcep=13,nfilt=26,nfft=512)
      x=(x-min)/(max-min)

      if(mode=='conv'):
        x=x.reshape(1,x.shape[0],x.shape[1],1)
      elif(mode=='time'):
        np.expand_dims(x,axis=0)
      y_hat=model.predict(x)
      y_prob.append(y_hat)
      y_pred.append(np.argmax(y_hat))
      y_true.append(c)
      
    fn_prob[fn]=np.mean(y_prob,axis=0).flatten()
  return(y_true,y_prob,fn_prob)
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/PGDBA-Project/V2/test_inst.csv")
```

```python
classes=list(np.unique(df.label))
fn2classes=dict(zip(df.fname,df.label))
```

```python
mode='conv'
#time
#min=-70.72182492951218
#max=69.9251863770808

#conv
min=-73.19197018743574
max=70.48366209252114
model = keras.models.load_model("/content/drive/MyDrive/Colab Notebooks/PGDBA-Project/V2/Conv_model.h5")
```

```python
y_true,y_pred,fn_prob=build_predictions("/content/drive/MyDrive/Colab Notebooks/PGDBA-Project/V2/clean_test/")
```

```python
y_probs=[]
for i,row in df.iterrows():
  y_prob=fn_prob[row.fname]
  y_probs.append(y_prob)
  for c,p in zip(classes,y_prob):
    df.at[i,c]=p
y_pred=[classes[np.argmax(y)] for y in y_probs]
df['y_pred']=y_pred
df.to_csv("/content/drive/MyDrive/Colab Notebooks/PGDBA-Project/V2/predictions.csv",index=False)
```

```python
acc_score=accuracy_score(y_true=df['label'],y_pred=df['y_pred'])
```
